```python
import joblib
import pandas as pd
import shap
import sys
sys.path.append("../src")

from preprocessing import load_data, clean_data

# Load model & preprocessor
model = joblib.load("../models/best_model.pkl")
preprocessor = joblib.load("../notebooks/data/processed/preprocessor.pkl")

# Load and clean raw data
df = load_data("../data/raw/telco_churn.csv")
df = clean_data(df)

X = df.drop("Churn", axis=1)
y = df["Churn"]

# Transform features
X_processed = preprocessor.transform(X)
```

```python
num_features = preprocessor.transformers_[0][2]
cat_features = preprocessor.transformers_[1][2]

ohe = preprocessor.transformers_[1][1]["onehot"]
ohe_features = ohe.get_feature_names_out(cat_features)

feature_names = list(num_features) + list(ohe_features)
```

```python
explainer = shap.Explainer(model, X_processed)
shap_values = explainer(X_processed)
```

```python
shap.summary_plot(
    shap_values,
    features=X_processed,
    feature_names=feature_names
)
```

High MonthlyCharges ↑ churn

Month-to-month contracts ↑ churn

Short tenure ↑ churn
